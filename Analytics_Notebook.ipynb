#### Skip the parts below, if you have already saved your data locally.

```python
# Load in the NFL Season 2020 data
YEAR = 2020


nfl2020_raw = pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                         'play_by_play_' + str(YEAR) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)
```

```python
# Clean the data by the redundant columns
nfl2020_raw.drop(['passer_player_name', 'passer_player_id',
           'rusher_player_name', 'rusher_player_id',
           'receiver_player_name', 'receiver_player_id'],
          axis=1, inplace=True)
```

```python
# Only use the regular season data
nfl2020 = nfl2020_raw.loc[nfl2020_raw.season_type=='REG']

# Remove irrelevant plays such as: kickoffs, field goals, kneel downs, etc.
nfl2020 = nfl2020.loc[(nfl2020.play_type.isin(['no_play','pass','run'])) & (nfl2020.epa.isna()==False)]

# Transform QB scrambles from run plays -> pass plays
nfl2020.play_type.loc[nfl2020['pass']==1] = 'pass'
nfl2020.play_type.loc[nfl2020.rush==1] = 'run'

nfl2020.reset_index(drop=True, inplace=True)
```

```python
# Saving the edited/cleaned data
nfl2020.to_csv('/Users/lukasjunger/PycharmProjects/football_webapp/football_analytics/data/nfl2020.csv.gz', compression='gzip', index=False)
```
